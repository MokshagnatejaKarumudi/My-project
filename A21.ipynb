```python
# Install necessary libraries
import joblib
from dash import Dash, dcc, html, Input, Output

# Load the model
model = joblib.load('model_joblib_gr')

# Initialize the Dash app
app = Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    html.H2("Rice Yield Prediction Model", style={'textAlign': 'center', 'color': '#4CAF50'}),

    dcc.Input(id='dist-code', type='number', placeholder='District Code', style={'margin': '10px'}),
    dcc.Input(id='year', type='number', placeholder='Year (2019-2031)', style={'margin': '10px'}),
    dcc.Input(id='rice-area', type='number', placeholder='Rice Area (ha)', style={'margin': '10px'}),
    dcc.Input(id='rice-production', type='number', placeholder='Rice Production (ton)', style={'margin': '10px'}),

    html.Button('Predict', id='predict-button', n_clicks=0, style={'margin': '10px'}),

    html.Div(id='result', style={'marginTop': '20px'})
])

# Define callback to update result based on user input
@app.callback(
    Output('result', 'children'),
    Input('predict-button', 'n_clicks'),
    Input('dist-code', 'value'),
    Input('year', 'value'),
    Input('rice-area', 'value'),
    Input('rice-production', 'value')
)
def predict_rice_yield(n_clicks, dist_code, year, rice_area, rice_production):
    if n_clicks > 0:
        if dist_code < 55 or dist_code > 63:
            return html.Span("Invalid District Code. Please enter a value between 55 and 63.", style={'color': '#FF5722'})
        elif year < 2019 or year > 2031:
            return html.Span("Invalid Year. Please enter a value between 2018 and 2030.", style={'color': '#FF5722'})
        else:
            prediction = model.predict([[dist_code, year, rice_area, rice_production]])
            return html.Span(f"Predicted Rice Yield: {prediction[0]:.2f}", style={'color': '#4CAF50'})

    return ""

# Run the app
if __name__ == '__main__':
    app.run_server(mode='inline')
```
